# Stock movement prediction


```python
# import library
import numpy as np
import pandas as pd
```

## data preprocess

```python
# read data
df = pd.read_csv('stock.csv')
df.Date = pd.to_datetime(df.Date)
print(df.columns)
df.head()
```

### new attribute

```python
# average line
# 
avg_line = []
LMean = []
SMean = []
df2 = df.copy().set_index('Date')
for date in df2.index:
    mean30 = df2.loc[date - pd.Timedelta('30D'): date]
    mean30 = mean30.Volume.mean()
    
    mean15 = df2.loc[date - pd.Timedelta('15D'): date]
    mean15 = mean15.Volume.mean()
    
    mean7 = df2.loc[date - pd.Timedelta('7D'): date]
    mean7 = mean7.Volume.mean()
  
    mean_long = mean15/mean30
    mean_short= mean7/mean15 
    
    LMean.append(mean_long)
    SMean.append(mean_short)
    """  
    if mean_short/mean_long>=1:
        if mean_long > 1 and mean_short > 1:
            avg_line.append(4)
        elif mean_long > 1 and mean_short<1:
            avg_line.append(3)
        elif mean_long < 1 and mean_short > 1:
            avg_line.append(2)
        elif mean_long < 1 and mean_short > 1:
            avg_line.append(1)
        else:
            avg_line.append(0)
    elif rate_short/rate_long<1:
        if mean_long > 1 and mean_short > 1:
            avg_line.append(-4)
        elif mean_long > 1 and mean_short<1:
            avg_line.append(-3)
        elif mean_long < 1 and mean_short > 1:
            avg_line.append(-2)
        elif mean_long < 1 and mean_short > 1:
            avg_line.append(-1)
        else:
            avg_line.append(0)
    """        
# df['Rate'] = avg_line
df['Long Mean'] = LMean
df['Short Mean'] = SMean
```

```python
# ratio 
open_rate = [0]
close_rate = [0]
high_rate = [0]
low_rate = [0]
for i in range(1, df.Date.size):
    open_rate.append(df['Open Price'][i] / df['Open Price'][i-1])
    close_rate.append(df['Close Price'][i] / df['Close Price'][i-1])
    high_rate.append(df['High Price'][i] / df['High Price'][i-1])
    low_rate.append(df['Low Price'][i] / df['Low Price'][i-1])
    
df['Open Rate'] = open_rate
df['Close Rate'] = close_rate
df['High Rate'] = high_rate
df['Low Rate'] = low_rate

df.head()
```

```python
# rise tomorrow
rise_1 = []
for i in range(0, df['Close Price'].size-1):
    if df['Close Price'][i+1] - df['Close Price'][i] >= 0:
        rise_1.append(1)
    else: rise_1.append(0)
rise_1.append(0)
df['Rise'] = rise_1
df.head()
```

### splitting testing/training data

```python
x_train = df[df.Date.dt.year <= 2017].drop(['Date', 'Open Price', 'Close Price', 'High Price', 'Low Price', 'Rise'], axis=1)
y_train = df[df.Date.dt.year <= 2017].Rise

x_test = df[df.Date.dt.year > 2017].drop(['Date', 'Open Price', 'Close Price', 'High Price', 'Low Price', 'Rise'], axis=1)
y_test = df[df.Date.dt.year > 2017].Rise

x_train.head()
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression as LR
clf = LR(random_state=0, solver='lbfgs')
clf = clf.fit(x_train, y_train)
clf.score(x_test, y_test)
```

## SVM

```python
from sklearn.svm import SVC
clf = SVC(gamma='scale')
clf = clf.fit(x_train, y_train)
clf.score(x_test, y_test)
```

## NN

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='lbfgs', alpha=0.5, hidden_layer_sizes=(5, 2), random_state=1)
print(clf)
clf = clf.fit(x_train, y_train)
clf.score(x_test, y_test)
```
